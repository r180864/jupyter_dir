# Data Visualization Using Matplotlib Assignment-3

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('cars-3.csv')
```

```python
df
```

```python
#1. Generating the Histogram
fig, ax=plt.subplots()
bins=[i for i in range(1, 31)]
ax.hist(df['wt'], bins=bins) #a, d. mapping wt x axis and setting bins
plt.xlabel('Weight of Cars') #b. labelling x-axis
plt.ylabel('Count') #c. labelling y-axis
plt.title('Histogram for the Weight Values') #e. setting the title
plt.show()
```
