# Data Visualization Using Matplotlib Assignment-1

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('cars-1.csv') #importing the data
```

```python
df
```

```python
#2. Generating the Line Plot Graph
plt.plot(df['model'], df['hp']) #a, b. mapping the model  and hp on the x and y axis respectively.
plt.xlabel('Models of the cars') #c. labelling x-axis
plt.ylabel('Horse-Power of cars') #d. Labelling y-axis
plt.title('Model Names vs Horse Power') #e. seting the title
plt.show() #Displaying the output
```
