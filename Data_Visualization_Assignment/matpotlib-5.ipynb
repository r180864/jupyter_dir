# Data Visualization Using Matplotlib Assignment-5

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('cars-5.csv')
```

```python
df
```

```python
#1. Plotting the area chart
plt.fill_between(df['am'], df['carb']) #a, b. setting the am and carb on x and y axes respectively
plt.xlabel('Transmission') #c. labelling x-axis
plt.ylabel('Number of Carburators') #d. Labelling y-axis
plt.title('Transmission vs Number of Carburators') #e. Probiding title
plt.show()
```
