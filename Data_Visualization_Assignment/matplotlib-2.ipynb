# Data Visualization Using Matplotlib Assignment-2

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('cars-2.csv')
```

```python
df
```

```python
#1. Generating the Bar Plot
plt.bar(df['carb'], df['gear']) #a, b. mapping crabs and gears on x and y axes respectively
plt.xlabel('Number of Carburators') #c. labelling x-axis
plt.ylabel('Number of Forward Gears') #d. labelling y-axis
plt.title('Carbs vs Gears') #e. setting the title
plt.show()
```
