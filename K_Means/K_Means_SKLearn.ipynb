```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
data=pd.read_excel('Dataset-Kmeans-xclara.xlsx')
```

```python
data.shape
```

```python
data.head()
```

```python
x=data['V1'].values
y=data['V2'].values
X=np.array(list(zip(x,y)))
```

```python
X
```

```python
kmeans=KMeans(n_clusters=3)
```

```python
kmeans=kmeans.fit(X)
```

```python
labels=kmeans.predict(X)
```

```python
centeroids=kmeans.cluster_centers_
```

```python
centeroids
```

```python
labels
```

```python
colors=["red","green","blue","yellow","pink","black","orange","purple","beige","brown","gray","cyan","magenta"]
fig, ax=plt.subplots()
for i in range(3):
    points=np.array([X[j] for j in range(len(X)) if labels[j]==i])
    ax.scatter(points[:, 0], points[:, 1], c=colors[i], s=7)
ax.scatter(centeroids[:, 0], centeroids[:, 1], marker='x', c='black', s=200)
```
