## Writing DataFrame to a file

```python
ids=pd.Series({'Anil': 'R180864', 'Sudheer':'R180340', 'Lokesh':'R181012', 'Rehaman':'R180603'})
branch=pd.Series({'Anil':'CSE', 'Rehaman':'CSE', 'Sudheer':'ECE', 'Sukumar':'ECE'})
surname=pd.Series({'Anil': 'Maddineni', 'Sudheer':'Poluru', 'Lokesh':'Talari', 'Rehaman':'Shaik', 'Sukumar':'Beesu'})
sec=pd.Series({'Anil': 'F001', 'Sudheer':'G008', 'Lokesh':'F006', 'Rehaman':'F008', 'Sukumar':'F001'})
students=pd.DataFrame({'Surname':surname, 'Id':ids, 'Branch':branch, 'Section': sec})
students
```

```python
#Writing DataFrame to a csv file
students.to_csv('students.csv')
pd.read_csv('students.csv')
```

```python
#To not add our indexes as a column in new file
#We will use the parameter index=False
students.to_csv('students.csv', index=False)
pd.read_csv('students.csv')
```

```python
#We cannot write Data to a particular sheet in a csv file
#Because csv file does not contain any sheets, it contain only a single sheet
students.to_csv('students.csv', sheet_name='details')
```

```python
#Writing DataFrame to an Excel file
students.to_excel('students.xlsx')
pd.read_excel('students.xlsx')
```

```python
#Writim=ng DataFrame to a specific sheet in anExcel file
students.to_excel('students.xlsx', sheet_name='details')
pd.read_excel('students.xlsx', sheet_name='details')
```

```python
#To not add our indexes as a column in new file
#We will use the parameter index=False
students.to_excel('students.xlsx', sheet_name='details', index=False)
pd.read_excel('students.xlsx', sheet_name='details')
```
