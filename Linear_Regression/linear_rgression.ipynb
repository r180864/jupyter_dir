SO there exist Linear Relationshop between these variables

```python
m=0
c=0
lr=0.0001
iterations=1000
error=[]
n=len(x)
for itr in range(iterations):
    error_cost=0
    cost_m=0
    cost_c=0
    for i in range(len(x)):
        y_pred=m*x[i]+c
        error_cost=(error_cost+(y[i]-y_pred)**2)/n
        for j in range(len(x)):
            partial_wrt_c=(2*((m*x[j]+c)-y[j]))/n
            partial_wrt_m=(2*((m*x[j]+c)-y[j])*x[j])/n
                       
            cost_m=cost_m+partial_wrt_m
            cost_c=cost_c+partial_wrt_c
                       
        m=m-lr*cost_m
        c=c-lr*cost_c
    #print(itr, c, m)
    error.append(error_cost)
#print('\nError Cost:\n', error)
mini=min(error)
print("\nMinimum Error:{}\nIteration:{}".format(mini, error.index(mini)))
print('C:', c)
print('M:', m)
```
