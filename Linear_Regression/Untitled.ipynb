```python
for col in df.columns:
    if(df[col].dtype=='object'):
        df[col]=df[col].fillna(df[col].mode()[0])
    else:
        df[col]=df[col].fillna(df[col].median())
```

```python
for col in df.columns:
    if(df[col].dtype!='object'):
        plt.hist(df[col])
        plt.xlabel(col)
        plt.ylabel("Count")
        plt.show()
```

```python
for col in df.columns:
    if(df[col].dtype!='object'):
        plt.boxplot(df[col])
        plt.xlabel(col)
        plt.ylabel("Count")
        plt.show()
```
