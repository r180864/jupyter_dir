# Predicting Rank based on Marks

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
marks['rank']=None
for i in range(len(marks)):
    marks['Rank'][i]=marks['Rank'][i][-1]
```
