# Tree

- **Tree** is a **non-linear** Data Structure which organizes data in **hierarchical** manner.
- Tree represents the nodes connected by edges.
- Tree originates from a particular node called **root** node, and continues with it's **child** nodes
- Each node has **one parent** only, but can have **multile children**.
- Each node is connected to its child via its **edge**.
- Tree can also be defined as a **"Collection of nodes(entities) linkd together to simulate a hierarchy"**.

## Basic Terminologies of Trees:

### 1. Node:
- Node is a structure which is going to store some informaiton and also containing link to the other nodes.
- Every node in a tree can have multiple or no child nodes to it.

### 2. Root:
- The top most element in a tree called Root.
- Root is this which is not having any parent.
- Every tree must have exactly one root node.

### 3. Edge:
- The connection between any two nodes is called edge.
- In any tree with N nodes, there will be maximum of N-1 edges.

### 4. Parent:
- The immediate predecessor of any node is known as parent of that node.
- Root is not having any parent node.

### 5. Child:
- The immediate successor of any node is known as child of that node.
- Leaf is not having any child node.

### 6. Siblings:
- All the childern of the same parent are known as siblings.

### 7. Leaf Node (External Node):
- The node which is having no child is known as leaf node.

### 8. Non-Leaf Node (Internal Node):
- The node which is having atleast one child is known as non-leaf node.
- All the other nodes except leaf nodes, are known as non-leaf nodes.

### 9. Degree:
- The no.of children of a node is known as degree of that particular node.
- Degree of all the leaf nodes is zero.
- The maximum degree of a node among all the nodes in a tree is known as degree of that tree.

### 10. Level:
- The distance from the root to a node is known as level of that node.
- No. of edges from root to given node, is known as level of that particular node.
 - Level of tree = Height of tree

### 11. Path:
- Sequence of connecting edges from source node to destination node is known as path between those nodes.

### 12. Height:
- No.of edges in the longest path from that node to a leaf node, is called of that node.
 - Height of a tree = Height of root node

### 13. Depth:
- Length of the path form root to that node, is known as depth of that node.
 - Depth of a tree = depth of leaf node

### 14. Ancestor:
- Any predecessor node on the path from root node to that node is called ancestor of that particular node.

### 15. Decendant:
- Any successor node on the path from that node to leaf node, is callled Decendant of that node.
