# Seaborn

```python
import seaborn as sns
sns.set()
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
#To read an excel file using seaborn
titanic=sns.load_dataset('titanic') #No need to specify extension like .xlsx, .cav, ...
#It will read the data in the format of Pandas DataFrame
```

```python
titanic.shape
```

```python
titanic.head()
```

## Dist Plot

```python
sns.distplot(titanic['age'])
```

```python
sns.distplot(titanic[['age']], bins=10)
```

## Pair plot

* It extracts only numerical data from the given DataFrame
* It wont take binary columns(True/False values) as input

```python
#These are all available numerical columns in DataFrame
titanic.describe()
```

```python
#We can also see the datatype of each column
titanic.dtypes
```

```python

titanic[['survived', 'pclass', 'sex', 'age', 'sibsp', 'parch',
       'fare', 'embarked', 'class', 'who', 'deck', 'embark_town',
       'alive']]
```

```python
#Passing only int and object type columns, without having boolean columns
sns.pairplot(titanic[['survived', 'pclass', 'sex', 'age', 'sibsp', 'parch',
       'fare', 'embarked', 'class', 'who', 'deck', 'embark_town',
       'alive']])
#It will consider only numerical values
```

## Count Plot
* It helps to capture the count of categorical variables

```python
sns.countplot(titanic['sex'])
plt.show()
```

```python
#To plot the graph horizontally
sns.countplot(y=titanic['survived'])
plt.show()
```

## Bar Plot
* Used for Bivariate Analysis
  * Whenever we will be having two variables included in computation

```python
titanic.head()
```

```python
sns.barplot(x='sex', y='fare', data=titanic)
plt.show()
```

```python
#The same thing as above can be acvhieved by this:
sns.barplot(x=titanic['sex'], y=titanic['fare'])
plt.show()
```

## Box Plot
* It is a robust ploy which will help us to deal with statistical parameters
  * Maximun capturing
  * Minimum Value
  * Quartiles
  * Median to be captured
* It also helps to identify the Outliers in the data

```python
sns.boxplot(titanic['age'])
```
