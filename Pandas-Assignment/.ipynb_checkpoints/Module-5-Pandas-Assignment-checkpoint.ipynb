```python
#2
def create_dataframe(keys, values):
    df=pd.DataFrame(values, index=keys)
    return df.T

keys=['One', 'Two']
values=[['X', 'Y'], ['A', 'B']]
m=create_dataframe(keys, values)
n=create_dataframe(keys, values) 
```

```python
#3
def add(df1, df2):
    df=pd.concat([df1, df2], axis=1, ignore_index=True)
    return df
```
