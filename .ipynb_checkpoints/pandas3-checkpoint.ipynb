```python
import pandas as pd
import numpy as np
import names
```

```python
WT=[np.random.randint(30, 100) for i in range(20)]
DSP=[np.random.randint(30, 100) for i in range(20)]
CD=[np.random.randint(30, 100) for i in range(20)]
COA=[np.random.randint(30, 100) for i in range(20)]
```

```python
students=[names.get_first_name() for i in range(20)]
```

```python
data=pd.DataFrame({"Name": students, "WT": WT, "DSP": DSP, "CD": CD, "COA": COA})
data
```

```python
dates=pd.date_range(start="2002-12-3", periods=20, freq='d')
dates
```

```python
data.index=dates
```

# Slicing and Indexing

1. Directly the columns
2. Uisng the LOC
3. Using the ILOC

###### Anytime we generate 1-D array in Pandas, it will give Series, as There are only 2 data structures in Pandas(Setries & DataFrame)

```python
grade=data
```

```python
#Accessing a single column as Series
name=grade["Name"]
print(type(name))
name
```

```python
#Accessing a Single colums as DataFrame by specifying column name in a list
name=grade[["Name"]]
print(type(name))
print(name)
```

```python
#We can't access rows directly
grade["2002-12-03"]
```

1. loc - Access columns and rows by it's name/value - End Position is Included
2. iloc(Index Location) - Access columns and names by its position - End Position is Excluded

```python
grade.loc[:, :]
```

```python
grade.loc[:"2002-12-15"]
```

```python
grade.loc[::2, "WT":"CD":2]
```
