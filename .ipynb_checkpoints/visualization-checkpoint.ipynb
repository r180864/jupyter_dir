# Color Codes
* k-black
* b-blue
* orange- orange
* m-magenta
* c-ayan
* y-yellow
* r-red
* g-green
* w-white

```python
plt.title("Height VS Weight Graph")
plt.plot(height, weight, 'g') #To plot the line in red color
plt.xlabel("Height")
plt.ylabel("Weight")
plt.show()
#We can use different colors like this
```

# Markers(ways to represent line)
* ^ - Upper Triangle
* v - inverted Triangle
* '>' - Right Side Triangle
* < - Left Side Triangle
* o - Circle
* '*' - Star
* d - Square
* -- - Dashed Line
* . - Dotted Line
* We can use combinations of these markers as well
#### We have to use these with color code

```python
plt.title("Height VS Weight Graph")
plt.plot(height, weight, "r--o")
plt.xlabel("Height")
plt.ylabel("Weight")
plt.show()
```

```python
#We can add different colors to line, marker face, marker edge separately with different colors
plt.title("Height VS Weight Graph")
plt.plot(height, weight, "r-o", mec= "k", mfc= "y")
#mec - Marker Edge Color
#mfc - Marker Face COlor
plt.xlabel("Height")
plt.ylabel("Weight")
plt.show()
```

```python
#We can also increase Marker SIze
plt.title("Height VS Weight Graph")
plt.plot(height, weight, "r-o", mec= "k", mfc= "y", ms='10') #ms - Marker Size
plt.xlabel("Height")
plt.ylabel("Weight")
plt.show()
```

```python
#To increase size of figure
plt.figure(figsize=(10,5))
plt.title("Height VS Weight Graph")
plt.plot(height, weight, "r-o", mec= "k", mfc= "y", ms="10")
plt.xlabel("Height")
plt.ylabel("Weight")
plt.savefig("mygrapg.jpg") #To save the grapg as image
plt.show()
```

```python
#To create grid
plt.figure(figsize=(10,5))
plt.title("Height VS Weight Graph")
plt.plot(height, weight, "r-o", mec= "k", mfc= "y", ms="10")
plt.xlabel("Height")
plt.ylabel("Weight")
plt.grid() #To create grid lines
plt.savefig("mygrapg.jpg") #To save the grapg as image
plt.show()
```

```python
#To set the limits for axes
plt.figure(figsize=(10,5))
plt.title("Height VS Weight Graph")
plt.plot(height, weight, "r-o", mec= "k", mfc= "y", ms="10")
plt.xlabel("Height")
plt.ylabel("Weight")
plt.xlim(50, 100)
plt.ylim(30, 100) #lim for limits
plt.show()
```

```python
#To set the intervals for the axes
plt.figure(figsize=(10,5))
plt.title("Height VS Weight Graph")
plt.plot(height, weight, "r-o", mec= "k", mfc= "y", ms="10")
plt.xlabel("Height")
plt.ylabel("Weight")
plt.xlim(50, 100)
plt.ylim(30, 100)
plt.xticks(np.arange(50, 101, 10)) #ticks for intervals
plt.yticks(np.arange(30, 101, 10))
plt.show()
```
